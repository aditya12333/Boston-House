```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
df = pd.DataFrame(boston.data, columns = boston.feature_names)
df.head()
```

```python
df['Price'] = pd.DataFrame(boston.target)
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
sns.scatterplot(x = "RM", y = "Price", alpha = 0.7, data = df)
```

```python
plt.figure(figsize = (12, 5))
sns.heatmap(df.corr(), annot = True)
```

```python
x = df.drop('Price', axis = 1)
y = df["Price"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

## test data

```python
pred = reg.predict(x_test)
```

```python
pred
```

```python
plt.scatter(y_test, pred, alpha = 0.7)
plt.xlabel('Predicet Price')
plt.ylabel('Actual Price')
plt.title("Actual Price vs Predicted Price")
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
score= r2_score(pred, y_test)
m = mean_absolute_error(pred, y_test)
ms = mean_squared_error(pred, y_test)
print('r2 value is', score)
print('mean absolute error id',  m)
print('mean squared error is',  ms)
```

## train data

```python
pred2 = reg.predict(x_train)
pred2
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
score= r2_score(pred2, y_train)
m = mean_absolute_error(pred2, y_train)
ms = mean_squared_error(pred2, y_train)
print('r2 value is', score)
print('mean absolute error id',  m)
print('mean squared error is',  ms)
```

the r2, MAE and MSE values of training data and test data are very close to eachother, so our model is not overfitted
